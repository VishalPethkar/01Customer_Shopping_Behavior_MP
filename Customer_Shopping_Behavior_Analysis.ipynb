```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(
    lambda x: x.fillna(x.median())
)

# Short Explanation
# 1. groupby('Category')
# --Groups the data based on each category.
# -- Example:
# Electronics
# Clothing
# Grocery



# 2. ['Review Rating']
# Selects only the Review Rating column.



# 3. transform(...)
# Applies an operation to every group and returns data in the original row size.



# 4. lambda x: x.fillna(x.median())
# -- For each category:
# finds the median review rating
# fills missing (NaN) values with that median


#imp -- “Fill missing review ratings using the median rating of their own category.”
```

```python
#create a column age_group

bins = [18,30,45,60,100]

labels = ['young adult','adult','middle-aged','senior']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)





# Explanation Step-by-Step
# 1. bins
# --bins = [18,30,45,60,100]

# These are age ranges:
# 18 → 30
# 30 → 45
# 45 → 60
# 60 → 100
# --Pandas uses these ranges to divide ages into groups.



# 2. labels
# --labels = ['young adult','adult','middle-aged','senior']
# --Names given to each age range.

# Age Range	Label
# 18–30	young adult
# 30–45	adult
# 45–60	middle-aged
# 60–100	senior



# 3. pd.cut()
# -- pd.cut(df['age'], bins=bins, labels=labels)
# -- pd.cut() converts continuous age values into categories.

# Example:
# age	Result
# 25	young adult
# 38	adult
# 52	middle-aged
# 70	senior



# 4. Create New Column
# --df['age_group'] = ...
# --Stores the result in a new column called age_group.

# Final Output Example
# age	age_group
# 22	young adult
# 35	adult
# 50	middle-aged
# 68	senior
```

```python
#create a column purchase_frequency_days


# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)





# Short Explanation
# 1.frequency_mapping
# --Creates a dictionary to convert text values into number of days.

# --Example:
# Weekly → 7
# Monthly → 30
# Annually → 365



# 2.map()
# --Matches values from frequency_of_purchases with the dictionary.



# 3.New Column
# --Creates a new column:
# purchase_frequency_days

# containing numeric day values.
```
